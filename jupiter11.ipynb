1. Importo la librería PANDAS para poder subir el archivo csv y leerlo para utilizar el  comando head con parámetro 10, que me permite imprimir las primeras 10 filas de la base de datos. 

```python
#Abrir librería PANDAS
import pandas as pd
#Importar datos
datos=pd.read_csv("A1.1 Obesidad.csv")
#Imprimir resumen de datos
print(datos.head(10))
```

3. Si quisiera realizar un estudio de inferencia, lo relacionaría con el peso, ya que pesos muy altos son garantizados de no entrar a la categoría de peso normal. 

```python
#Importar numpy
import numpy as np
promedio=np.mean(datos.Peso)
print("El promedio del peso es: ", promedio)
```

4. Importo la librería para realizar la gráfica, en este caso escogí el grupo de mujeres y decidí comparar la edad con el peso. 

```python
#Importar librería
import matplotlib.pyplot as plt
#Función mágica
%matplotlib inline
plt.scatter(x = datos.Edad[datos.Sexo == "Female"], y = datos.Peso[datos.Sexo == "Female"], alpha = 0.05)
plt.show()
```
